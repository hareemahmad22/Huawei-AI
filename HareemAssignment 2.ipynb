```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 3, 5, 7, 9])
y = np.array([250, 350, 450, 550, 650])
n = len(x)
```

```python
w = 0  # initial weight (slope)
b = 0  # initial bias (intercept)
alpha = 0.01  # learning rate
epochs = 1000  # number of iterations
```

```python
for i in range(epochs):
    y_pred = w * x + b             # predicted values
    error = y_pred - y             # difference between predicted and actual

    dw = (1/n) * np.dot(error, x)  # gradient w.r.t. weight
    db = (1/n) * np.sum(error)     # gradient w.r.t. bias

    w -= alpha * dw  # update weight
    b -= alpha * db  # update bias

    # Optional: Print progress every 100 steps
    if i % 100 == 0:
        cost = (1/(2*n)) * np.sum(error**2)
        print(f"Epoch {i}: Cost={cost:.2f}, w={w:.4f}, b={b:.4f}")
```

```python
print("\nFinal model parameters:")
print(f"Weight (slope): {w:.4f}")
print(f"Bias (intercept): {b:.4f}")
```

```python
y_hat = w * x + b
print("\nPredicted values:", y_hat)
```

```python
plt.scatter(x, y, color='blue', label='Actual')   # original data
plt.plot(x, y_hat, color='red', label='Predicted')  # regression line
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid(True)
plt.show()
```
